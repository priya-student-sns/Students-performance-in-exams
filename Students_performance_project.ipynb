```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.histplot(df['math score'])
plt.title("visualising the math score")
plt.show()
```

```python
sns.boxplot(x='gender', y='writing score', data=df)
plt.title("visualising the writing score ")
plt.show()
```

```python
sns.barplot(x='test preparation course',y='math score',data=df)
plt.title("visualising the math score")
plt.show()
```

```python
sns.pairplot(df[['math score','writing score','reading score']])
plt.suptitle("visualising the pairs of dataset" ,y=1.2)
plt.show()
```

```python
df.groupby('test preparation course')[['math score','writing score','reading score']].mean()
```
